```python
import re
from swifter import swifter
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
#1. Cleaning
def _cleaning(hasil):
    # Hapus NaN
    hasil = hasil.dropna()

    # Hapus mention
    hasil['cleaning'] = hasil['Text Tweet'].str.replace(r"@[A-Za-z0-9_]+", "", regex=True)
    # Hapus hashtag
    hasil['cleaning'] = hasil['cleaning'].str.replace(r"#[A-Za-z0-9_]+", "", regex=True)
    # Hapus URL
    hasil['cleaning'] = hasil['cleaning'].str.replace(r'http\S+|www.\S+', '', regex=True)
    # Hapus seluruh kata yang mengandung angka, contoh: tiba2
    hasil['cleaning'] = hasil['cleaning'].str.replace(r'\d+', '', regex=True)
    # Hapus seluruh karakter yang tidak termasuk alphabet
    hasil['cleaning'] = hasil['cleaning'].str.replace(r'[^a-zA-Z]', ' ', regex=True)
    # Menghilangkan kata dengan huruf ganda, contoh: paketttt
    hasil['cleaning'] = hasil['cleaning'].str.replace(r'(.)\1+', r'\1', regex=True)
    # Hapus karakter satu huruf
    hasil['cleaning'] = hasil['cleaning'].str.replace(r'\b[a-zA-Z]\b', '', regex=True)
    # Hapus double space
    hasil['cleaning'] = hasil['cleaning'].str.replace(r' +', ' ', regex=True)

    # Hapus NaN
    hasil = hasil.dropna()
    return hasil

#2. Case folding
def _case_folding(hasil):
    hasil['komentar_casefolding'] = hasil['cleaning'].str.lower()
    return hasil

#3. Tokenizing
def _tokenizing(hasil):
    hasil['komentar_tokenizing'] = hasil['komentar_casefolding'].str.split()
    return hasil

#4. Stopword
def _stopword(hasil):
    stop_factory = StopWordRemoverFactory().get_stop_words()

    hasil['komentar_stopword'] = hasil['komentar_tokenizing'].swifter.apply(lambda x: ([word for word in x if word not in (stop_factory)]))
    return hasil

#5. Stemming
def _stemming(hasil):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    hasil['komentar_stemming'] = hasil['komentar_stopword'].swifter.apply(lambda x: ([stemmer.stem(word) for word in x]))
    return hasil
```

```python
df = _cleaning(df)
df = _case_folding(df)
df = _tokenizing(df)
df = _stopword(df)
df = _stemming(df)

df
```

```python
from sklearn.model_selection import train_test_split
# Ambil data teks dan label dari dataframe
X = df['komentar_stemming']
y = df['Sentimen']

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train.shape)
print(X_test.shape)
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def identity_function(text):
    return text

tfidf_vectorizer = TfidfVectorizer(max_features=1500, analyzer=identity_function)   #analyzer untuk tfidf dapat menerima input dalam bentuk token

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
pd.DataFrame(X_train_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

### SMOTE

```python
pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train_tfidf, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import Counter

# Ambil distribusi sebelum dan setelah SMOTE secara otomatis
before_smote = Counter(y_train)  # Distribusi sebelum SMOTE
after_smote = Counter(y_train_resampled)  # Distribusi setelah SMOTE

# Buat DataFrame dengan format yang sesuai
df_smote = pd.DataFrame({
    "Kondisi": ["Sebelum SMOTE"] * 2 + ["Setelah SMOTE"] * 2,  # Sumbu X
    "Jumlah": [before_smote['negative'], before_smote['positive'], after_smote['negative'], after_smote['positive']],  # Sumbu Y
    "Kategori": ["Negatif", "Positif"] * 2  # Hue (warna)
})

# Buat plot menggunakan Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x="Kondisi", y="Jumlah", hue="Kategori", data=df_smote, palette=["#FF9999", "#66B3FF"])

# Tambahkan label dan judul
plt.xlabel("Kondisi Data")
plt.ylabel("Jumlah Data")
plt.title("Distribusi Data Sebelum dan Setelah SMOTE")
plt.legend(title="Kategori")

# Tampilkan plot
plt.show()
```

### Training model Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

# Train MultinomialNB
model = MultinomialNB().fit(X_train_resampled, y_train_resampled)
y_pred = model.predict(X_test_tfidf)
```

```python
# Fungsi untuk preprocessing teks baru sebelum prediksi
def preprocess_text(text):
    # Cleaning
    text = re.sub(r"@[A-Za-z0-9_]+", "", text)  # Hapus mention
    text = re.sub(r"#[A-Za-z0-9_]+", "", text)  # Hapus hashtag
    text = re.sub(r'http\S+|www.\S+', '', text)  # Hapus URL
    text = re.sub(r'\d+', '', text)  # Hapus angka
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # Hapus karakter selain huruf
    text = re.sub(r'(.)\1+', r'\1', text)  # Hapus huruf ganda
    text = re.sub(r'\b[a-zA-Z]\b', '', text)  # Hapus karakter satu huruf
    text = re.sub(r' +', ' ', text).strip()  # Hapus spasi berlebih

    # Case folding
    text = text.lower()

    # Tokenizing
    tokens = text.split()

    # Stopword removal
    stop_factory = StopWordRemoverFactory().get_stop_words()
    tokens = [word for word in tokens if word not in stop_factory]

    # Stemming
    stemmer = StemmerFactory().create_stemmer()
    tokens = [stemmer.stem(word) for word in tokens]

    return tokens  # Output dalam bentuk token untuk TF-IDF

# Contoh kalimat untuk diuji
dt_pred = "layanan pengiriman ini termasuk Lambat banget"

# Preprocessing teks uji
dt_pred_clean = preprocess_text(dt_pred)

# Transformasi ke TF-IDF
dt_pred_tfidf = tfidf_vectorizer.transform([dt_pred_clean])

# Prediksi Sentimen
my_pred = model.predict(dt_pred_tfidf)

# Menampilkan hasil
print("Komentar :", dt_pred)
print("Komentar bersih :", dt_pred_clean)
print("Sentimen : ", my_pred[0])
```
